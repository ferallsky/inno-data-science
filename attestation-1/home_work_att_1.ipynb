Сделаем несколько предположений о зависимости в данных:
1. В зимнее время потребление должно быть больше
2. Так как на западе страны располагается крупнейший промышленный центр Индии Мумбаи, то западный регион должен возглавлять список самых "прожорливых"

```python
# Выведем среднее потребление
avg_consumption = data.mean().sort_values(ascending=False).reset_index().rename(columns = {"index": "state", 0 : "avg_consumption"})
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x= "state", y = "avg_consumption", data = avg_consumption)
plt.xticks(rotation=70,fontweight='light', fontsize='x-large')
plt.show()
```

```python
# Сгруппируем данные по регионам и посмотрим среднее потребление
data_region = pd.DataFrame()
data_region['NR'] = data['Punjab']+ data['Haryana']+ data['Rajasthan']+ data['Delhi']+data['UP']+data['Uttarakhand']+data['HP']+data['J&K']+data['Chandigarh']
data_region['WR'] = data['Chhattisgarh']+data['Gujarat']+data['MP']+data['Maharashtra']+data['Goa']+data['DNH']
data_region['SR'] = data['Andhra Pradesh']+data['Telangana']+data['Karnataka']+data['Kerala']+data['Tamil Nadu']+data['Pondy']
data_region['ER'] = data['Bihar']+data['Jharkhand']+ data['Odisha']+data['West Bengal']+data['Sikkim']
data_region['NER'] = data['Arunachal Pradesh']+data['Assam']+data['Manipur']+data['Meghalaya']+data['Mizoram']+data['Nagaland']+data['Tripura']
data_region['Date'] = data.index

data_region_line = pd.DataFrame({"Северный регион": data_region["NR"].values,
                        "Южный регион": data_region["SR"].values,
                        "Восточный регион": data_region["ER"].values,
                        "Западный регион": data_region["WR"].values,
                        "Северо-восточный регион": data_region["NER"].values},index=data_region.Date)

data_region_line
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data = data_region_line)
plt.show()
```

```python
from IPython.utils.text import string
# Посмотрим распределение по месяцам
data['Date'] = pd.to_datetime(data.index, format='%Y%m')
months = data.groupby(data['Date'].dt.strftime('%B')).sum().sort_values(by='Date')
#avg_consumption_months = months.mean().sort_values(ascending=False).reset_index().rename(columns = {"index": "state", 0 : "avg_consumption"})
months
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data = months)
plt.show()
```

```python
data_region_line['Date'] = pd.to_datetime(data_region_line.index, format='%Y%m')
data_region_months = data_region_line.groupby(data_region_line['Date'].dt.strftime('%B')).sum().sort_values(by='Date')
#avg_consumption_months = months.mean().sort_values(ascending=False).reset_index().rename(columns = {"index": "state", 0 : "avg_consumption"})
data_region_months
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data = data_region_months)
plt.show()
```

2. Наиболее загруженные месяцы - это Январь, февраль, март, Апрель. Чуть ниже идет Май. Это можно как раз объяснить тем, что зима в Индии попадает в период с декабря по начало апреля.
